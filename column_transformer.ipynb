```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2)
```

## Aam Zindagi

```python
## we will fill nan value
from sklearn.impute import SimpleImputer
si=SimpleImputer()
```

```python
X_train_fever=si.fit_transform(X_train[['fever']])
X_test_fever=si.transform(X_test[['fever']])
```

```python
X_train_fever.shape
```

```python
X_test_fever.shape
```

```python
X_train.head(5)
X_train['city'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder,OrdinalEncoder
```

```python
#Ordinalencoding the fever column
oe=OrdinalEncoder(categories=[['Mild','Strong']])
```

```python
X_train_cough=oe.fit_transform(X_train[['cough']])
X_test_cough=oe.transform(X_test[['cough']])
X_test_cough.shape
```

```python
##one hot encodding of gender,city
from sklearn.preprocessing import OneHotEncoder
```

```python
oeh=OneHotEncoder(drop='first',sparse_output=False)
X_train_gender_city=oeh.fit_transform(X_train[['gender','city']])
X_test_gender_city=oeh.transform(X_test[['gender','city']])
X_train_gender_city.shape
```

```python
##extracing  age from X-train and X-test
X_train_age=X_train.drop(columns=['gender','fever','cough','city']).values
X_test_age=X_test.drop(columns=['gender','fever','cough','city']).values
X_test.shape
```

```python
## merging the Columns
X_train_transformed=np.concatenate((X_train_age,X_train_fever,X_train_cough,X_train_gender_city),axis=1)
X_test_transformed=np.concatenate((X_test_age,X_test_fever,X_test_cough,X_test_gender_city),axis=1)
X_train_final=pd.DataFrame(X_train_transformed)
X_test_final=pd.DataFrame(X_test_transformed)

```

## Mentos Zindagi

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

```python
X_train.head(3)
```

```python
Transformer=ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]) , ['cough']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first') , ['gender','city'])
],remainder='passthrough')
output=Transformer.fit_transform(X_train)
```
