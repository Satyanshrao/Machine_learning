```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('titanic_mixed_Variables.csv')
```

```python
df['number'].unique()
```

```python
## extract Numerical part
df['number_numerical']=pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
## extract Categorical part
df['number_categorical']=np.where(df['number_numerical'].isnull(),df['number'],np.nan)
```

```python
df.head(5)
```

```python
df['Cabin'].unique()
```

```python
df['Cabib_number']=df['Cabin'].str.extract('(\d+)')
```

```python
df['Cabin_category']=df['Cabin'].str[0]
```

```python
df.head(5)
```

```python
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
# Splits the ticket string by spaces
# Takes the last part → usually the number
df['ticket_num'] = pd.to_numeric(df['ticket_num'], errors='coerce', downcast='integer')
# Converts it to numeric
# Invalid values → NaN
# Uses smaller integer type (memory efficient)
# 🔹 Part 2: Extract first part of ticket (category)
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
# Takes the first part of the ticket → usually a prefix/category (like "PC", "A", etc.)
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan, df['ticket_cat'])
```
