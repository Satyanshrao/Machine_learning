```python
import pickle
import numpy as np
import pandas as pd
```

```python
ohe_sex=pickle.load(open('models/ohe_sex.pkl','rb'))
ohe_embarked=pickle.load(open('models/ohe_embarked.pkl','rb'))
clf=pickle.load(open('models/clf.pkl','rb'))

```

```python
## data feeding
## rem/age/embarked/sex
test_input=np.array([1,0,0,10.5,31.0,'S','male'],dtype='object').reshape(1,7)
```

```python
test_input
```

```python
sex_df = pd.DataFrame(test_input[:, -1], columns=['Sex'])
test_input_sex=ohe_sex.transform(sex_df)
```

```python
test_input_sex
```

```python
test_input_embarked=ohe_embarked.transform(test_input[:,5].reshape(1,1))
```

```python
test_input_embarked
```

```python
test_input_age=test_input[:,4].reshape(1,1)
```

```python
test_input_age
```

```python
test_input_transformed=np.concatenate((test_input[:,[0,1,2,3]],test_input_age,test_input_embarked,test_input_sex),axis=1)
```

```python
test_input_transformed.shape
```

```python
clf.predict(test_input_transformed)
```
